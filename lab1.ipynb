```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Створимо певний набір даних з інформацією про відсоток безробіття протягом року
data = {
    'Місяць': ['Січень', 'Лютий', 'Березень', 'Квітень', 'Травень', 'Червень', 'Липень', 'Серпень', 'Вересень', 'Жовтень', 'Листопад', 'Грудень'],
    'Рівень безробіття (%)': [8.1, 7.9, 7.8, 7.6, 7.5, 7.3, 7.4, 7.2, 7.0, 6.9, 6.8, 6.7]
}
```

```python
# Створимо DataFrame
df = pd.DataFrame(data)
```

```python
# Побудова лінійного графіку
plt.figure(figsize=(10, 6))
plt.plot(df['Місяць'], df['Рівень безробіття (%)'], marker='o')
plt.title('Рівень безробіття протягом року')
plt.xlabel('Місяць')
plt.ylabel('Рівень безробіття (%)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
